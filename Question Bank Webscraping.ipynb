## Webscraping to get placement aptitude questions from a website

```python
#Importing the libraries
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

#Creating a driver
options = webdriver.ChromeOptions()
options.add_argument("headless")
driver = webdriver.Chrome("C:\Program Files\Google\Chrome\Application\chromedriver.exe")

#Creating lists to store values temporarily
questions=[] #List to store questions
answers=[] #List to store all the options
qno = []#List to store question number
ano = []#List to store correct answer's option value

#Guiding the driver to open the website
driver.get("https://www.geeksforgeeks.org/probability-gq/")
content = driver.page_source
soup = BeautifulSoup(content)

#Parsing all the questions
for ques in soup.findAll('div', attrs={'class':'mtq_question_text'}):
    questions.append(ques.text)

#Parsing all the answers
for ans in soup.findAll('div', attrs={'class':'mtq_answer_text'}):
    answers.append(ans.text)

#Parsing the correct answer details    
id_var = []
for res in soup.findAll('div', attrs={'class':'mtq_marker mtq_correct_marker'}):
    temp = res.parent
    temp = str(temp)
    temp = temp.split('"')
    id_var.append(temp[7])

#Separating all the options into their own category
o1 = []
o2 = []
o3 = []
o4 = []
for i in id_var:
    i = i.split('-')
    #Obtaining the question number and the correct option
    qno.append(i[1])
    ano.append(i[2])


lst = np.array_split(answers, 10)
for i in range(0,10):
    o1.append(lst[i][0])
    o2.append(lst[i][1])
    o3.append(lst[i][2])
    o4.append(lst[i][3])
    
#Storing all the data into a csv file
df = pd.DataFrame({'Question Number':qno,'Problem':questions,'Option 1':o1,'Option 2':o2,'Option 3':o3,'Option 4':o4,'Correct answer':ano})
df.to_csv('qbank.csv', index=False, encoding='utf-8')

#Reading the newly created file
df = pd.read_csv('qbank.csv')
df
```
